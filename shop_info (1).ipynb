```python
import bs4
from bs4 import BeautifulSoup
import requests
```

```python
url="https://yellowpages.org.in/books-And-publishers/22"
```

Sending a GET request to the specified URL to retrieve the webpage content.

```python
page=requests.get(url)
page
```

UParse the HTML content of the webpage into a structured format using BeautifulSoup, and prettify the HTML for easier readability.

```python
soup=BeautifulSoup(page.text)
soup.prettify
```

Find all HTML div elements with the class "listing-info listpopular" in the parsed webpage, which likely contains the publisher information.

```python
publisher=soup.find_all("div",{"class":"listing-info listpopular"})
publisher
```

Extract all HTML h3 elements with the class "title" from the parsed webpage, which likely contains the article titles.

```python
titles=soup.find_all("h3",{"class":"title"})
titles
```

Find all HTML li elements with the class "place-location" in the parsed webpage, which likely contains the location information.

```python
locs=soup.find_all("li",{"class":"place-location"})
locs
```

Initialize empty lists to store the extracted publisher names and locations, respectively.

```python
publisher_names=[]
locations=[]
```

Iterate through the extracted title elements, extract the publisher name from each title's anchor tag, strip any unnecessary whitespace, and append it to the publisher_names list, then print the list of extracted publisher names.

```python
for title in titles:
    publisher_name=title.find('a').text.strip()
    publisher_names.append(publisher_name)
print(publisher_names)
```

Iterate through the extracted location elements, extract the location text from each location's anchor tag, strip any unnecessary whitespace, and append it to the locations list, then print the list of extracted locations.

```python
for loc in locs:
    loc_list=loc.find('a').text.strip()
    locations.append(loc_list)
print(locations)
```

The list of extracted location information.

```python
locations
```

```python
publisher_names
```

Create a dictionary to store the extracted data, with keys "publisher_names" and "locations" mapping to the corresponding lists of extracted data.

```python
Data={"publisher_names":publisher_names,"locations":locations}
```

Import the pandas library

```python
import pandas as pd
```

converting dictionary "Data" into pandas dataframe

```python
shop_info=pd.DataFrame(Data)
```
